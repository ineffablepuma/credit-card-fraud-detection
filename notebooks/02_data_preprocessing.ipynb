```python
import pandas as pd
df = pd.read_csv('../data/raw/fraudTest.csv')

# Quick look
df.head()
```

```python
import pandas as pd

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

# Create additional time features
df['trans_hour'] = df['trans_date_trans_time'].dt.hour
df['trans_day'] = df['trans_date_trans_time'].dt.day
df['trans_weekday'] = df['trans_date_trans_time'].dt.weekday
df['trans_month'] = df['trans_date_trans_time'].dt.month

# Calculate age
df['age'] = (pd.to_datetime('today') - df['dob']).dt.days // 365
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
cols = ['merchant', 'category','gender', 'job','cc_num','street','city','state','zip',]

df[cols].nunique()
```

```python
#columns to drop
cols_to_drop = [
    # Index / IDs
    'Unnamed: 0',
    'cc_num',
    'trans_num',

    # Personal info
    'first',
    'last',
    'street',
    'dob',

    # Raw datetime & time representations
    'trans_date_trans_time',
    'unix_time',

    # Location text & codes
    'zip',
    'city',

    # Raw latitude & longitude (after distance feature)
    'lat',
    'long',
    'merch_lat',
    'merch_long'
]

# Drop all at once
df = df.drop(columns=cols_to_drop)

# Verify
print(df.columns)
print(df.shape)
```

```python
df.head()
```

```python
df.duplicated().sum()
#df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
merchant_freq = df['merchant'].value_counts(normalize=True)
df['merchant_enc'] = df['merchant'].map(merchant_freq)
df = df.drop(columns=['merchant'])
#Captures how often a merchant appears
```

```python
df.head()
```

```python
job_freq = df['job'].value_counts(normalize=True)
df['job_enc'] = df['job'].map(job_freq)
df = df.drop(columns=['job'])
```

```python
df.head()
```

```python
#after doing frequency encoding which job has more fraud is not important but if the job is rare or not is important.
```

```python
df = pd.get_dummies(df, columns=['category'], drop_first=True)
#does one hot encoding creating 14 diff columns, category is a important factor in fraud.
```

```python
df.head()
```

```python
df['gender'] = df['gender'].map({'M': 0, 'F': 1})
```

```python
df.info()
```

```python
# Convert all boolean columns to int
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
state_freq = df['state'].value_counts(normalize=True)
df['state_enc'] = df['state'].map(state_freq)
df = df.drop(columns=['state'])
# frequency encoding for 50states
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['amt', 'city_pop', 'age', 'merchant_enc', 'job_enc',
            'trans_hour', 'trans_day', 'trans_weekday', 'trans_month']

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.to_csv('../data/processed/cleaned_data.csv', index=False)
```
